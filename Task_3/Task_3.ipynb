### There is higher sales and profit of Consumer and the Home Office has the least sales and profit in relation to segments

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['yellow','red'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

### Technology category has the highest sale and profit and Furniture category has least sales and proft in relation to Category of product

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['blue','green'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

```python
plt.figure(figsize=(20,10))
df.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(color=['red','blue'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

### The PHONES and CHAIRS has same number of sales but Phones creates more profits than chairs
### Fasterner has least sales with no profit
### bookcases and tables has good sales but there is loss

```python
df.groupby('Ship Mode')[['Profit','Sales']].sum().plot.bar(color=['orange','lightblue'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

### There is higher sales when products are shipped using Standard class ship mode
